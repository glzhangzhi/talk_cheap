进程 process  multiprocess
线程 thread   threading

一个应用程序至少包括一个进程，而一个进程至少包括一个线程

进程之间的内存单元独立，而一个进程的多个线程之间共享内存

理论上来说，一个cpu只能运行一个进程，在单cpu的情况下多进程是通过在不同进程之前快速切换形成的效果。而多核cpu的出现使真正的多进程成为可能。

python中，对于Cpython解释器，存在GIL锁，即一个进程中，永远只有一个线程可以拿到资源，而此时如果强行使用多线程，会出现当一个GIL被释放时，多个线程争夺资源，线程之间互相切换，从而浪费资源。现代的GIL会在程序休眠时自动释放GIL，所以如果用time.sleep()来测试多线程，表现应该是与多进程相同的。

对于CPU密集型，多进程更快，对于IO密集型，多线程更快
但实验效果来说，无脑用多进程，除非在做网络请求密集型任务，CPU闲置率过高，此时用多进程很浪费CPU资源，应考虑使用多线程。
